```python
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i["name"])
    return L
```

```python
movie['genres']=movie['genres'].apply(convert)
```

```python
movie['keywords']=movie['keywords'].apply(convert)
```

```python
def convert1(obj):
    L=[]
    obj=ast.literal_eval(obj)
    for i in range(0,3):
        if i<len(obj):
            L.append(obj[i]['name'])
    return L
```

```python
movies['cast']=movies['cast'].apply(convert1)
```

```python
movies.sample(10)
```

```python
movie['production_companies']=movie['production_companies'].apply(convert)
movie['production_countries']=movie['production_countries'].apply(convert)
movie['spoken_languages']=movie['spoken_languages'].apply(convert)
```

```python
movies.sample(15)
```

```python
def convert2(obj):
    L=[]
    obj=ast.literal_eval(obj)
    for i in obj:
        if ((i['job']=='Director')|(i['job']=='Producer')):
            L.append(i['name'])
    return L
```

```python
movies['crew']=movies['crew'].apply(convert2)
```

```python
movies.sample(10)
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.sample(15)
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
movies['production_companies']=movies['production_companies'].apply(lambda x:[i.replace(" ","") for i in x])
movies['production_countries']=movies['production_countries'].apply(lambda x:[i.replace(" ","") for i in x])
movies['spoken_languages']=movies['spoken_languages'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.sample(15)
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['production_companies']+movies['production_countries']+movies['spoken_languages']+movies['cast']+movies['crew']
```

```python
movies=movies[['movie_id','title','tags']]
```

```python
movies.sample(10)
```

```python
movies['tags']=movies['tags'].apply(lambda x:" ".join(x))
```

```python
movies.sample(10)
```

```python
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()

def stem(text):
    y = []

    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
movies['tags']=movies['tags'].apply(stem)
```

```python
movies.sample()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
vectors=cv.fit_transform(movies['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors)
```

```python
similarity
```

```python
def recommend(movie):
    movie_index=movies[movies['title']==movie].index[0]
    distances=similarity[movie_index]
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(movies.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
